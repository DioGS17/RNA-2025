```python
columns_num=['Elevation','Aspect', 'Slope', 'Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology',
            'Horizontal_Distance_To_Roadways', 'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm',
            'Horizontal_Distance_To_Fire_Points']

columns_soil=['Soil_Type1', 'Soil_Type2', 'Soil_Type3', 'Soil_Type4', 'Soil_Type5', 'Soil_Type6', 'Soil_Type7',
              'Soil_Type8', 'Soil_Type9', 'Soil_Type10', 'Soil_Type11', 'Soil_Type12', 'Soil_Type13', 'Soil_Type14',
              'Soil_Type15', 'Soil_Type16', 'Soil_Type17', 'Soil_Type18', 'Soil_Type19', 'Soil_Type20', 'Soil_Type21',
              'Soil_Type22', 'Soil_Type23', 'Soil_Type24', 'Soil_Type25', 'Soil_Type26', 'Soil_Type27', 'Soil_Type28',
              'Soil_Type29', 'Soil_Type30', 'Soil_Type31', 'Soil_Type32', 'Soil_Type33', 'Soil_Type34', 'Soil_Type35',
              'Soil_Type36', 'Soil_Type37', 'Soil_Type38', 'Soil_Type39', 'Soil_Type40',]

columns_wild=['Wilderness_Area1', 'Wilderness_Area2', 'Wilderness_Area3', 'Wilderness_Area4']
```

```python
df.hist(figsize=(12, 16), column=columns_num, grid=False)
plt.plot()
```

```python
cols = 2
rows = (len(columns_num)+1) // 2

fig, axes = plt.subplots(rows, cols, figsize=(10, 24))
axes = axes.flatten()

for i, var in enumerate(columns_num):
    sns.boxplot(data=df, y=var, width=0.2, ax=axes[i])
    axes[i].set_xlabel(f'{var}')
    axes[i].set_ylabel('')
```

```python
df['Soil_Type'] = df[columns_soil].idxmax(axis=1).str.extract(r'(\d+)').astype(int)
df['Wilderness_Area'] = df[columns_wild].idxmax(axis=1).str.extract(r'(\d+)').astype(int)
```

```python
soil_counts = df['Soil_Type'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.barplot(x=soil_counts.index, y=soil_counts.values)
plt.title('Distribuição dos Tipos de Solo (Soil_Type)')
plt.xlabel('Tipo de Solo')
plt.ylabel('Frequência')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
wild_counts = df['Wilderness_Area'].value_counts().sort_index()

plt.figure(figsize=(6, 4))
sns.barplot(x=wild_counts.index, y=wild_counts.values)
plt.title('Distribuição das Áreas Selvagens (Wilderness_Area)')
plt.xlabel('Área Selvagem')
plt.ylabel('Frequência')
plt.show()
```

```python
df = df.drop(columns=['Soil_Type', 'Wilderness_Area'])
```

```python
cover_type_counts = df['Cover_Type'].value_counts().sort_index()

sns.barplot(x=cover_type_counts.index, y=cover_type_counts.values)
plt.title('Distribuição dos Tipos de Cobertura (Cover_Type)')
plt.xlabel('Tipo de Cobertura')
plt.ylabel('Frequência')
plt.show()
```

Analisando o gráfico de barras do atributo alvo, percebe-se que ele é muito desbalanceado, tendo os tipos 1 e 2 acima de 200 mil exemplos e o resto abaixo de 50 mil.

Como o atributo alvo assume múltiplos valores, o tipo de tarefa de ML é multiclasse.

Algumas das métricas compatíves para este problema são:
* Acurácia
* Matriz de confusão
* Revocação, Precisão e F1-score para cada classe
* Macro F1-score e Weighted F1-score

Como o atributo alvo é desbalanceado, usaremos como referência as métricas:
* Macro F1-score: para poder avaliar o desempenho do modelo de forma justa entre as classes
* Weighted F1-score: para avaliar o desempenho "geral" do modelo, considerando o desbalanceamento entre as classes
